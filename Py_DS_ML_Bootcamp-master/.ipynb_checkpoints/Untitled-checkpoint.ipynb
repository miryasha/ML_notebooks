```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('./bull.csv')

df.head()
```

```python
# Select the two columns you want to analyze
column1 = df['signal_date_num_day']
column2 = df['tp_date_num_day']

# Calculate the correlation between the two columns
correlation = column1.corr(column2)

# Print the correlation
print("The correlation between column1 and column2 is:", correlation)

# Plot a scatter plot to visualize the correlation
plt.scatter(column1, column2)
plt.xlabel('signal_date_num_day')
plt.ylabel('tp_date_num_day')
plt.title('Scatter plot of Column 1 and Column 2')
plt.show()
```
